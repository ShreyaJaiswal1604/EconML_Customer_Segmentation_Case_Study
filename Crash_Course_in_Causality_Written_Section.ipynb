Method 1: Randomized Experiments

Imagine you want to know if showing an ad makes people more likely to buy a game. The best way to find out is like this:

Purchase Game
See Advertisement
Now, the best way to get a clear answer is to do a special test, like a game experiment:

You gather a bunch of friends who like games.
You randomly pick some friends to see a cool ad for a new game (Treatment group).
Others don't see any ad (Control group).
This is like flipping a coin to decide who sees the ad. Here's why it's cool:

Since it's random, there's no reason certain types of friends see the ad; it's like luck.
So, any difference you see between friends who saw the ad and those who didn't is probably because of the ad, not because of other things (like maybe they already love games a lot).
It's like saying, "Let's play a fair game to see if this ad really makes a difference in getting people to buy the game." That's why it's called the gold standard - it's a really good way to find out the real reason behind things.

Q. what if by luck the firends who see the add are actually game lovers?

That's a great point! If, by luck, the friends who see the ad happen to be game lovers more often, it could mess up our results. That's because we wouldn't be sure if the ad is making people buy games or if it's just because the friends who already love games were the ones who saw the ad.

This is why even though randomizing helps, it's essential to have a large enough group of friends so that, on average, the two groups (those who see the ad and those who don't) are similar. This helps balance out any accidental differences.

However, in real life, it's not always possible to control every factor, and sometimes, we may need to be aware that our results could be influenced by these factors. That's why scientists and researchers try their best to design experiments carefully and analyze the results critically.


---

Let's break it down with an example:

Imagine you have 100 friends.
You use a fair method to randomly select 50 friends to see the ad (Group A) and 50 friends not to see the ad (Group B).
Now, the goal is to make sure that, on average, the friends in Group A are the same as those in Group B in terms of things that might affect whether they buy games or not (like how much they already love games).

If, by chance, Group A has more game lovers than Group B, any differences in game purchases between the two groups might be because of their existing love for games, not just the ad.

By having a large enough group and using randomization, we hope that, on average, any accidental differences between the groups will balance out. This way, we can be more confident that any differences we observe in game purchases are likely due to the ad and not to other factors.


---
does this mean having tw groups without any bias?
Yes, exactly! When we say the two groups are similar, we mean that they are created without any intentional or systematic bias. The goal is to have both groups as comparable as possible, ensuring that any differences observed between them can be attributed to the treatment (in this case, seeing or not seeing the ad) and not to other factors.

Randomization is a key tool in achieving this. By randomly assigning individuals to either the treatment group (those who see the ad) or the control group (those who don't see the ad), we aim to eliminate any systematic biases or pre-existing differences between the two groups. This helps ensure that any observed effects are more likely to be caused by the treatment rather than by other variables.

Random Assignment:

Random assignment is a method used in experiments where individuals are randomly assigned to different groups, such as a treatment group (receives a particular intervention or exposure) and a control group (does not receive the intervention). The idea is to distribute both known and unknown variables equally across the groups, creating comparable groups and minimizing the influence of confounding variables.

Pros:

Gold Standard for Causal Effect:

Random assignment is considered the gold standard for isolating the causal effect of a treatment. Since individuals are randomly assigned to groups, any observed differences in outcomes between groups can be attributed to the treatment rather than pre-existing differences.
No Need to Identify Confounding Variables:

In a well-designed randomized experiment, there's no need to identify or measure all possible confounding variables. The random assignment process helps in balancing these variables across the groups, making the treatment group and the control group comparable.
Cons:

Impractical for Some Treatments:

For certain treatments, conducting randomized experiments may be impractical or cost-prohibitive. For example, it might be challenging or unethical to randomly assign people to smoking or non-smoking groups in a study on the health effects of smoking.
Non-compliance:

In some cases, individuals assigned to a particular group may not comply with their assignment. For example, if a customer is offered a small discount but repeatedly visits the site in the hope of getting a larger discount, it can disrupt the intended experimental design.
Lack of Guaranteed Participation:

Even if customers are encouraged to join a loyalty program, there is no guarantee that they will actually join. This introduces the potential for non-compliance, where individuals assigned to the treatment group may not follow through with the intended action.
In summary, while random assignment is a powerful tool for experimental design, it is not always feasible or appropriate for every research question. Researchers must carefully consider the nature of the treatment and the practical constraints of the study when deciding whether to use random assignment or alternative research methods.

Method 2: Measure Confounders

Imagine you want to know if showing an ad makes people more likely to buy a game. However, you know that some factors (confounders) might affect both the likelihood of seeing the ad and buying the game. For example, people who spend a lot of time on gaming websites might both see the ad more often and be more likely to buy games.

Method 2: Measure Confounders in Action:

Instead of relying on random assignment like in a randomized experiment, you decide to carefully measure and consider these factors (confounders) in your study. So, you gather information about how much time each person spends on gaming websites.
Full Observability (Unconfoundedness):

If you can accurately measure and account for all the factors that might influence both seeing the ad and buying the game, your statistical models can then separately estimate each part of the process. This situation, where you can measure all the confounding influences, is known as "unconfoundedness."
Example:

Let's say you find out that, indeed, people who spend more time on gaming websites are more likely to see the ad and more likely to buy games. However, when you carefully account for this in your analysis, you may find that the ad still has a positive effect on game purchases, even after considering the influence of time spent on gaming websites.
So, in simple terms, Method 2 involves measuring and considering all the factors that could mess up your study (confounders) to get a clear understanding of the real impact of the treatment (ad) on the outcome (game purchase). This is known as unconfoundedness when you can observe and account for everything that might influence the results.

---

Observed Confounder Estimation in Action:
Instead of running a special experiment, you decide to use real-world data. You collect information about customers' characteristics, like how much time they spend on gaming websites, their age, and other relevant details.

Pros:

This method helps you figure out if the ad really influences game purchases without needing a fancy experiment or tool.
It's like detective work with data - you're trying to untangle the mystery of whether the ad is the real reason behind more game purchases.
Cons:

The challenge is that you need to identify and measure all the things that might affect both seeing the ad and buying the game. For instance, if people who already love games are more likely to see the ad and also more likely to buy games, you need to account for this in your analysis. If you miss something important, your conclusion might be off.
So, observed confounder estimation in the ad and customer example is like being a detective with a checklist. You need to name and measure everything that might mess up your study to make sure you understand if the ad is truly making people buy more games.

Method 3: Instrumental Variables

Scenario:

You want to figure out if a loyalty program (treatment) makes customers more likely to respond to advertisements and make purchases. However, you can't randomly assign customers to the loyalty program because you want it to be fair.

Instrumental Variables (IV) in Action:

Instrumental Variable (IV):

Imagine you discover that the proximity of a customer's home to the store randomly influences whether they join the loyalty program. Customers who live closer are more likely to join because it's convenient, and those who live farther away are less likely.
Using IV Estimation:

Instead of directly controlling who joins the loyalty program, you use the distance to the store as an instrumental variable. It's like a natural factor that influences whether customers join the program. Now, you can analyze how changes in distance (the instrumental variable) affect loyalty program enrollment, ad responsiveness, and purchase behavior.
Pros:

This method works even if there are other factors (confounders) you can't measure or control that might influence both loyalty program enrollment and purchasing behavior.
It's like using the distance to the store as a natural trick to help you separate the impact of the loyalty program on ad responsiveness and purchases.
Example Clarification:

If, after accounting for the distance to the store, you still see that customers in the loyalty program are more likely to respond to ads and make purchases, you can be more confident that the loyalty program is causing these effects, rather than other factors related to the distance.
So, in simple terms, using instrumental variables in the customer and ad example is like finding a natural element (distance to the store) that helps you understand if a loyalty program really influences customer behavior, even when you can't control all the factors involved.
